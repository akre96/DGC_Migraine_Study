```python
import numpy as np
import stumpy
import matplotlib.pyplot as plt
from scipy.signal import square
from matplotlib.patches import Rectangle
```

```python
# Sin Wave
sine_period=100
sine_ts = np.sin(np.arange(1000)*(2*np.pi/sine_period))

# Square Wave
sq_period=100
square_ts = square(np.arange(1000)*(2*np.pi/sq_period))

# Step Function
step_ts = np.array([0]*500 + [1]*500, dtype=np.float64)

time_series = sine_ts + square_ts
window_size = 50  # Approximately, how many data points might be found in a pattern

matrix_profile = stumpy.stump(time_series*10, m=window_size)
```

```python
mp = matrix_profile
```

```python
view_window=200

fig, axs = plt.subplots(2, sharex=True, gridspec_kw={'hspace': 0}, figsize=(16,8))
plt.suptitle('Motif (Pattern) Discovery', fontsize='30')

axs[0].plot(time_series[:view_window])
axs[0].set_ylabel('Original', fontsize='20')

axs[1].set_xlabel('Time', fontsize ='20')
axs[1].set_ylabel('Matrix Profile 0', fontsize='20')

axs[1].plot(mp[:view_window, 0])

```
